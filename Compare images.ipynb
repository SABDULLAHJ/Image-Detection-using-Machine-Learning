```python
# import the necessary packages
from skimage.measure import structural_similarity as ssim
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def mse(image1,image2):
    err = np.sum((image1.astype(float)-image2.astype(float))**2)
    err /= float(image1.shape[0] * image2.shape[1])
    return err
```

```python
def compare_images(image1,image2,title):
    m = mse(image1,image2)
    s= ssim(image1,image2)
    
    #setup the figure
    fig = plt.figure(title)
    plt.suptitle("MSE : %.2f, SSIM : %.2F" %(m,s))
    
    
    # show first image
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(image1, cmap = plt.cm.gray)
    plt.axis("off")

    # show the second image
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(image2, cmap = plt.cm.gray)
    plt.axis("off")

    # show the images
    plt.show()
```

```python
# load the images -- the original, the original + contrast,
# and the original + photoshop
original = cv2.imread("C:\\Users\\Syed Abdullah Jelani\\Desktop\\##\\python-compare-two-images\\images\jp_gates_original.png")
contrast = cv2.imread("C:\\Users\\Syed Abdullah Jelani\\Desktop\\##\\python-compare-two-images\\images\jp_gates_contrast.png")
shopped = cv2.imread("C:\\Users\\Syed Abdullah Jelani\\Desktop\\##\\python-compare-two-images\\images/jp_gates_photoshopped.png")
```

```python
# convert the images to grayscale
original = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
contrast = cv2.cvtColor(contrast, cv2.COLOR_BGR2GRAY)
shopped = cv2.cvtColor(shopped, cv2.COLOR_BGR2GRAY)
```

```python
# initialize the figure
fig = plt.figure("Images")
images = ("Original", original), ("Contrast", contrast), ("Photoshopped", shopped)
```

```python
for (i, (name,image)) in enumerate(images):
    ax = fig.add_subplot(1, 3, i + 1)
    ax.set_title(name)
    plt.imshow(image, cmap = plt.cm.gray)
    plt.axis("off")
```

```python
# show the figure
plt.show()
```

```python
# compare the images
compare_images(original, original, "Original vs. Original")
compare_images(original, contrast, "Original vs. Contrast")
compare_images(original, shopped, "Original vs. Photoshopped")
```
